```python
import pandas as pd
import plotly.express as px
df = pd.read_csv("/Users/Anushka/Desktop/HackHers2024/transactions_v2.csv")
```

```python
def top_n_categories(df, column_name, n):
    counts = df[column_name].value_counts()
    top_n_values = counts.head(n)
    return top_n_values.index.tolist()



top_3_categories = top_n_categories(df, 'category', 3)
print("Top 3 most frequent categories in 'category' column:", top_3_categories)
```

```python
def top_n_categories(df, column_name, n):
    category_sums = df.groupby(column_name)['amount'].sum()  # Summing up the amounts for each category
    top_n_categories = category_sums.nlargest(n)  # Selecting the top n categories with the highest sums
    return top_n_categories.index.tolist()

top_3_categories = top_n_categories(df, 'category', 3)
print("Top 3 categories with the highest sums in 'category' column:", top_3_categories)
```

```python
def top_n_names(df, column_name, n):
    name_counts = df[column_name].value_counts()  # Count occurrences of each name
    top_n_names = name_counts.head(n)  # Select top n names with the highest counts
    return top_n_names.index.tolist()

top_3_names = top_n_names(df, 'name', 3)
print("Top 3 most frequent names in 'name' column:", top_3_names)
```

```python
fig = px.bar(df, x='category', y='amount', title='Amount Spent by Category', color_discrete_map={'category': 'red'})
fig.show()
```
